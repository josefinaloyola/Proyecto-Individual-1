```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
amazon_df = pd.read_csv('amazon_prime_titles.csv')

disney_plus_df = pd.read_csv('disney_plus_titles.csv')

hulu_df = pd.read_csv('hulu_titles.csv')

netflix_df = pd.read_csv('netflix_titles.csv')
```

```python
#arreglo el tipo de datos
# amazon
amazon_df['show_id'] = amazon_df['show_id'].astype('str') # porque hay que arreglar el tipo de dato?
amazon_df['type'] = amazon_df['type'].astype('category')
amazon_df['title'] = amazon_df['title'].astype('str')
amazon_df['director'] = amazon_df['director'].astype('str')
# amazon_df['cast'] = amazon_df['cast'].astype('str')
amazon_df['country'] = amazon_df['country'].astype('str')
amazon_df['date_added'] = pd.to_datetime(amazon_df['date_added'], infer_datetime_format=True, errors='coerce')
amazon_df['release_year'] = amazon_df['release_year'].astype('int16')
# amazon_df['rating'] = amazon_df['rating'].astype('str')
amazon_df['listed_in'] = amazon_df['listed_in'].astype('str')
amazon_df['description'] = amazon_df['description'].astype('str')
```

```python
# disney_plus
disney_plus_df['show_id'] = disney_plus_df['show_id'].astype('str') # porque hay que arreglar el tipo de dato?
disney_plus_df['type'] = disney_plus_df['type'].astype('category')
disney_plus_df['title'] = disney_plus_df['title'].astype('str')
disney_plus_df['director'] = disney_plus_df['director'].astype('str')
# disney_plus_df['cast'] = disney_plus_df['cast'].astype('str')
disney_plus_df['country'] = disney_plus_df['country'].astype('str')
disney_plus_df['date_added'] = pd.to_datetime(disney_plus_df['date_added'], infer_datetime_format=True, errors='coerce')
disney_plus_df['release_year'] = disney_plus_df['release_year'].astype('int16')
# disney_plus_df['rating'] = disney_plus_df['rating'].astype('str')
disney_plus_df['listed_in'] = disney_plus_df['listed_in'].astype('str')
disney_plus_df['description'] = disney_plus_df['description'].astype('str')
```

```python
# hulu
hulu_df['show_id'] = hulu_df['show_id'].astype('str') # porque hay que arreglar el tipo de dato?
hulu_df['type'] = hulu_df['type'].astype('category')
hulu_df['title'] = hulu_df['title'].astype('str')
hulu_df['director'] = hulu_df['director'].astype('str')
# hulu_df['cast'] = hulu_df['cast'].astype('str')
hulu_df['country'] = hulu_df['country'].astype('str')
hulu_df['date_added'] = pd.to_datetime(hulu_df['date_added'], infer_datetime_format=True, errors='coerce')
hulu_df['release_year'] = hulu_df['release_year'].astype('int16')
# hulu_df['rating'] = hulu_df['rating'].astype('str')
hulu_df['listed_in'] = hulu_df['listed_in'].astype('str')
hulu_df['description'] = hulu_df['description'].astype('str')
```

```python
# netflix
netflix_df['show_id'] = netflix_df['show_id'].astype('str') # porque hay que arreglar el tipo de dato?
netflix_df['type'] = netflix_df['type'].astype('category')
netflix_df['title'] = netflix_df['title'].astype('str')
netflix_df['director'] = netflix_df['director'].astype('str')
# netflix_df['cast'] = netflix_df['cast'].astype('str')
netflix_df['country'] = netflix_df['country'].astype('str')
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'], infer_datetime_format=True, errors='coerce')
netflix_df['release_year'] = netflix_df['release_year'].astype('int16')
# netflix_df['rating'] = netflix_df['rating'].astype('str')
netflix_df['listed_in'] = netflix_df['listed_in'].astype('str')
netflix_df['description'] = netflix_df['description'].astype('str')
```

```python
# 1) Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets
amazon_df['id'] = 'a' + amazon_df['show_id']
amazon_df['id'] = amazon_df['id'].astype('str')
```

```python
disney_plus_df['id'] = 'd' + disney_plus_df['show_id']
disney_plus_df['id'] = disney_plus_df['id'].astype('str')
```

```python
hulu_df['id'] = 'h' + hulu_df['show_id']
hulu_df['id'] = hulu_df['id'].astype('str')
```

```python
netflix_df['id'] = 'n' + netflix_df['show_id']
netflix_df['id'] = netflix_df['id'].astype('str')
```

```python
# 2) Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity rating: “general for all audiences”)

amazon_df['rating'].fillna('G', inplace= True)
disney_plus_df['rating'].fillna('G', inplace=True)
hulu_df['rating'].fillna('G', inplace=True)
netflix_df['rating'].fillna('G', inplace=True)

hulu_df.head(5)
```

```python
# 3)  De haber fechas, deberán tener el formato **`AAAA-mm-dd`**
amazon_df['date_added'] = pd.to_datetime(amazon_df['date_added'], infer_datetime_format=True, errors='coerce')

# using errors=’coerce’. It will replace all non-numeric values with NaN

disney_plus_df['date_added'] = pd.to_datetime(disney_plus_df['date_added'], infer_datetime_format=True, errors='coerce')
hulu_df['date_added'] = pd.to_datetime(hulu_df['date_added'], infer_datetime_format=True, errors='coerce')
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'], infer_datetime_format=True, errors='coerce')
hulu_df.head(5)
```

```python
# 5) El campo ***duration*** debe convertirse en dos campos: **`duration_int`** y **`duration_type`**.
# 5) El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

amazon_df[['duration_int', 'duration_type']] = amazon_df['duration'].str.split(' ', n=-1, expand=True, regex=None)
amazon_df = amazon_df[amazon_df.columns[:10].tolist() + ['duration_int', 'duration_type'] + amazon_df.columns[10:-2].tolist()]

disney_plus_df[['duration_int', 'duration_type']] = disney_plus_df['duration'].str.split(' ', n=-1, expand=True, regex=None)
disney_plus_df = disney_plus_df[disney_plus_df.columns[:10].tolist() + ['duration_int', 'duration_type'] + disney_plus_df.columns[10:-2].tolist()]

hulu_df[['duration_int', 'duration_type']] = hulu_df['duration'].str.split(' ', n=-1, expand=True, regex=None)
hulu_df = hulu_df[hulu_df.columns[:10].tolist() + ['duration_int', 'duration_type'] + hulu_df.columns[10:-2].tolist()]

netflix_df[['duration_int', 'duration_type']] = netflix_df['duration'].str.split(' ', n=-1, expand=True, regex=None)
netflix_df = netflix_df[netflix_df.columns[:10].tolist() + ['duration_int', 'duration_type'] + netflix_df.columns[10:-2].tolist()]

hulu_df.head(5)
```

```python
# eliminar columna duration
amazon_df.drop('duration', axis=1, inplace=True)
disney_plus_df.drop('duration', axis=1, inplace=True)
hulu_df.drop('duration', axis=1, inplace=True)
netflix_df.drop('duration', axis=1, inplace=True)

hulu_df.head(5)
```

```python
#uno las tablas a un df multiplataforma
df_multiplat= pd.concat([amazon_df,disney_plus_df,hulu_df,netflix_df])
```

```python
#4) Pongo todo en minusculas, sin excepciones
df_multiplat['id'] = df_multiplat['id'].str.lower()
df_multiplat['type'] = df_multiplat['type'].str.lower()
df_multiplat['title'] = df_multiplat['title'].str.lower()
df_multiplat['director'] = df_multiplat['director'].str.lower()
df_multiplat['cast'] = df_multiplat['cast'].str.lower()
df_multiplat['country'] = df_multiplat['country'].str.lower()
df_multiplat['rating'] = df_multiplat['rating'].str.lower()
df_multiplat['listed_in'] = df_multiplat['listed_in'].str.lower()
df_multiplat['description'] = df_multiplat['description'].str.lower()
df_multiplat['duration_type'] = df_multiplat['duration_type'].str.lower()

df_multiplat.head(5)
```

```python
#Creo el df con las tablas de ratings
score1 = pd.read_csv('rating1.csv')
score2 = pd.read_csv('rating2.csv')
score3 = pd.read_csv('rating3.csv')
score4 = pd.read_csv('rating4.csv')
score5 = pd.read_csv('rating5.csv')
score6 = pd.read_csv('rating6.csv')
score7 = pd.read_csv('rating7.csv')
score8 = pd.read_csv('rating8.csv')
```

```python
df_scores= pd.concat([score1, score2, score3, score4, score5, score6, score7, score8])
```

```python
#Renmobro columnas id para que sea equivalente nombre del otro df, y tmb columna rating a score para que sea distinta a columna del otro df.
df_scores.rename(columns = {'movieId':'id'}, inplace = True)
df_scores.rename(columns = {'rating':'score'}, inplace = True)
```

```python
#Calculo el score por película xq hay varios usuarios que dan puntaje a la misma peli
scores_mean = df_scores.groupby(['id']).mean()
scores_mean.head(5)
```

```python
# Elimino columnas userId y timestamp que no me sirven
scores_mean.drop('userId', axis=1, inplace=True)
scores_mean.drop('timestamp', axis=1, inplace=True)
```

```python
#Uno la tabla scores con la tabla de plataformas
df_final = pd.merge(df_multiplat, scores_mean, on="id")
```

```python
# Seteo en 0 los shows que tienen duracion NaN

df_final['duration_int'].fillna(0, inplace = True)
df_final['duration_int'] = df_final['duration_int'].astype('int')
```

```python
# Película con mayor duración con filtros opcionales de AÑO, PLATAFORMA Y TIPO DE DURACIÓN
# la función debe llamarse get_max_duration(year, platform, duration_type)

#no tiene sentido agregar filtro de tipo de duración porque todas las películas están seteadas en minutos. El otro tipo de duración es season/s y solo aplica a series

def get_max_duration(plataforma, year:int):
        dicc = {"netflix":"n","amazon":"a","hulu":"h","disney_plus":"d"}
        
        if (plataforma!="todas"): 
                if year!=0:
                        df_temp = df_final[(df_final['id'].str[0] == dicc[plataforma])&(df_final['type'] == 'movie')&(df_final['release_year'] == year)].sort_values(["duration_int"],ascending=[False]).reset_index(drop=True)
        
                        titulo = df_temp.iloc[0,2]
                        duracion = df_temp.iloc[0,9]
                        return f' La pelicula de mayor duración en la plataforma {plataforma} el año {year} es {titulo} , con una duracion de {duracion}'
        
                elif year == 0:
                        df_temp = df_final[(df_final['id'].str[0] == dicc[plataforma])&(df_final['type'] == 'movie')].sort_values(["duration_int"],ascending=[False]).reset_index(drop=True)
        
                        titulo = df_temp.iloc[0,2]
                        duracion = df_temp.iloc[0,9]
                        return f' La pelicula de mayor duración en la plataforma {plataforma} es {titulo} , con una duracion de {duracion}'
        
        elif (plataforma=="todas"): 
                if year!=0:
                        df_temp = df_final[(df_final['type'] == 'movie')&(df_final['release_year'] == year)].sort_values(["duration_int"],ascending=[False]).reset_index(drop=True)
        
                        titulo = df_temp.iloc[0,2]
                        duracion = df_temp.iloc[0,9]
                        return f' La pelicula de mayor duración en el año {year} es {titulo} , con una duracion de {duracion}'
        
                elif year == 0:
                        df_temp = df_final[(df_final['type'] == 'movie')].sort_values(["duration_int"],ascending=[False]).reset_index(drop=True)
        
                        titulo = df_temp.iloc[0,2]
                        duracion = df_temp.iloc[0,9]
                        return f' La pelicula de mayor duración es {titulo} , con una duracion de {duracion} minutos'
        
```

```python
get_max_duration("hulu",2019)

# Si se pasa "todas" como parámetro quiere decir que se están eligiendo todas las plataformas
# Si se pasa 0 como parámetro quiere decir que no se está eligiendo ningun año
```

```python
# Cantidad de películas por plataforma con un puntaje mayor a XX en determinado año 
# la función debe llamarse get_score_count(platform, scored, year)

def get_score_count(plataforma:str, point:int, anio:int):

    dicc = {"netflix":"n",
            "amazon":"a",
            "hulu":"h",
            "disney_plus":"d"}
    df_temp = df_final[(df_final["score"]>point) & 
                        (df_final["release_year"]==anio) & 
                        (df_final.id.str[0] == dicc[plataforma]) & 
                        (df_final["type"] == "movie")]
    
    cantidad = len(df_temp.index) 

    return f' Hay {cantidad} peliculas con un puntaje mayor a {point} en {plataforma} en el año {anio}'
```

```python
get_score_count("disney_plus", 2, 2019)
```

```python
# Cantidad de películas por plataforma con filtro de PLATAFORMA. 
# La función debe llamarse get_count_platform(platform)

def get_count_platform(platform):
    dicc = {"netflix":"n",
              "amazon":"a",
              "hulu":"h",
              "disney_plus":"d"}
    df_temp = df_final[(df_final.id.str[0] == dicc[platform]) & (df_final["type"] == "movie")]
    
    cantidad = len(df_temp.index)
    
    return f' Hay {cantidad} peliculas en {platform}'
```

```python
get_count_platform("hulu")
```

```python
# Actor que más se repite según plataforma y año. 
# La función debe llamarse get_actor(platform, year))

def get_actor(plataforma, anio):
    
    dicc = {"netflix":"n",
            "amazon":"a",
            "hulu":"h",
            "disney_plus":"d"}
    
    # Elimino todos los valores nulos de la columna cast
    df_final.dropna(subset = ['cast'], inplace = True)
    
    df_temp = df_final[(df_final["release_year"]==anio) & (df_final.id.str[0] == dicc[plataforma]) & (df_final["type"] == "movie")]

    new_df = df_temp['cast'].value_counts(ascending=False)    
    
    actor = new_df.index[0]
    cantidad = new_df[0]
   
    
    return f"El actor de {plataforma} que más se repite es {actor} apareciendo un total de {cantidad} veces en el año {anio}"
    
```
